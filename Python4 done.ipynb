````python
data = input("Введите количество раскрытых дел в каждом месяце, разделенных пробелом: ")

data = data.replace(',', '')

crimes = tuple(map(int, data.split()))

max_cases = 0
min_cases = 100

for cases in crimes:
    if cases > max_cases:
        max_cases = cases
    if cases < min_cases:
        min_cases = cases

print("Максимальное количество дел в месяц:", max_cases)
print("Минимальное количество дел в месяц:", min_cases)



````

В программе дана заготовка функции, которая должна принимать на вход целое положительное число n и возвращать n-ое число последовательности Фибоначчи, используя цикл for. Дополните заготовку в указанных местах так, чтобы функция выполняла поставленную задачу. Программа должна запрашивать ввод с клавиатуры целого положительного числа n и выводить на экран результат работы вашей функции — n-ное число последовательности Фибоначчи.

***СПРАВОЧНО:*** Числа Фибоначчи — числовая последовательность, первые два элемента которой — 0 и 1, а каждое последующее число равно сумме двух ему предшествующих.

### Пример

**Ввод:**

```
10
```

**Вывод:**

```
55
```

````python
# дополните пропуски в этой ячейке
def fibon(num):
    num1 = num2 = 1
    if num < 2:
        return num1
    for i in range(2, num):
        num1, num2 = num2, num1 + num2
    return num2

````

````python
n = int(input("Введите целое положительное число n: "))
result = fibon(n)
print("n-ое число последовательности Фибоначчи:", result)
````

## Задача 6*. Алгоритм Евклида

Алгоритм Евклида для поиска НОД.

НОД – наибольший общий делитель, – число, делящее без остатка пару чисел и являющееся наибольшим из возможных делителей. Например: НОД(8, 6)=2; НОД(12, 8)=4. Очень важное понятие, необходимость в вычислении которого возникает достаточно часто. Простой расчетной формулы для НОД не существует. Решение полным перебором возможно, но для больших чисел необходимо большое количество вычислительных операций. Хорошее решение проблемы поиска наибольшего общего делителя двух чисел нашел еще Евклид. В самом простом случае алгоритм Евклида применяется к паре положительных целых чисел и формирует новую пару, которая состоит из меньшего числа и разницы между большим и меньшим числом. Процесс повторяется, пока числа не станут равными. Найденное число и есть наибольший общий делитель исходной пары.

Напишите функцию NOD, которая принимает на вход два целых числа и возвращает для них наибольший общий делитель.

### Пример

**Ввод:**

```
32
24
```

**Вывод:**

```
8
```

````python
def NOD(a, b):
    while b:
        a, b = b, a % b
    return a


````

````python
num1 = int(input("Введите первое целое число: "))
num2 = int(input("Введите второе целое число: "))
result = NOD(num1, num2)
print("Наибольший общий делитель:", result)
````
